```python
file_path1 = '/content/drive/My Drive/Colab Notebooks/Asgmt3/Population.csv'
file_path2 = '/content/drive/My Drive/Colab Notebooks/Asgmt3/Metadata_Country.csv'
```

```python
import pandas as pd

# Load the population dataset, specifying the encoding
pop_Data = pd.read_csv(file_path1) # or 'cp1252', or other encodings
Country_meta_Data = pd.read_csv(file_path2) # Apply the same encoding for other files if needed
```

```python
pop_Data.head()
```

```python
pop_Data.columns
```

```python
# Columns to retain
columns_to_keep = ['Country Name', 'Country Code', '2020', '2021', '2022', '2023']

# Create a new dataframe with only the selected columns
pop_Data_subset = pop_Data[columns_to_keep]

# Inspect the subset to ensure it only contains the selected columns
print("Subset of Population Data with selected columns:")
pop_Data_subset.head()
```

```python
# Check for missing value in Population subset
print("Population subset dataset checksum:")
print(pop_Data_subset.isnull().sum())
```

```python
print("\nCountry Metadata Cleaned:")
Country_meta_Data.head()
```

```python
print(Country_meta_Data.isnull().sum())
```

```python
# Drop unnecessary columns in the metadata country dataset
Country_meta_Data = Country_meta_Data.drop(columns=['TableName', 'Unnamed: 5', 'SpecialNotes'], errors='ignore')

# Fill missing values for 'Region', 'IncomeGroup', and 'SpecialNotes'
Country_meta_Data['Region'] = Country_meta_Data['Region'].fillna('Unknown')
Country_meta_Data['IncomeGroup'] = Country_meta_Data['IncomeGroup'].fillna('Unknown')

# Inspect the cleaned dataset
print("After filling missing values:")
Country_meta_Data.head()
```

```python
print(Country_meta_Data.isnull().sum())
```

```python
# Merge datasets on 'Country Code'
merged_population_data = pd.merge( Country_meta_Data, pop_Data_subset, on='Country Code', how='inner')

# Inspect the merged dataset
print("Merged Data:")
merged_population_data.head()
```

```python
merged_population_data.shape
```

```python
# Save the merged data
merged_population_data.to_csv('merged_population_data.csv', index=False)

# If you're using Google Colab and need to download it:
from google.colab import files
files.download('merged_population_data.csv')
```
